```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_df = pd.read_csv('/content/train.csv')
test_df    = pd.read_csv('/content/test.csv')

# preview the data
titanic_df.head()
```

```python
titanic_df.info()
print("----------------------------")
test_df.info()
```

```python
titanic_df.isnull().sum()
```

```python
# Drop rows with missing values in the "Embarked" column.
titanic_df.dropna(subset=["embarked"], inplace=True)

# Fill missing values in the "Cabin" column with the string "Unknown".
#titanic_df["Cabin"].fillna("Unknown", inplace=True)

# Fill missing values in the "Age" column with the mean of the "Age" column.
titanic_df["age"].fillna(titanic_df["age"].mean(), inplace=True)
```

```python
plt.figure(figsize=(6, 3))
sns.histplot(titanic_df["age"], kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.countplot(data = titanic_df, x="sex", hue="survived")
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="survived", loc="upper right")
plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.scatterplot(data= titanic_df , x="age", y="fare", hue="survived")
plt.title("Scatter Plot of Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title="Survived")
plt.show()
```
